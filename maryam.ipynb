```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e_x / e_x.sum(axis=-1, keepdims=True)

# Self-attention function
def compute_self_attention(sentence):
    tokens = sentence.split()
    n_tokens = len(tokens)
    d_k = 64

    # Random embeddings for demonstration
    np.random.seed(42)
    embeddings = np.random.randn(n_tokens, d_k)

    # Q, K, V
    Q = embeddings
    K = embeddings
    V = embeddings

    # Step 1: Dot product
    scores = np.matmul(Q, K.T)

    # Step 2: Scaling
    scaled_scores = scores / np.sqrt(d_k)

    # Step 3: Softmax
    attention_weights = softmax(scaled_scores)

    return tokens, attention_weights

# Input sentence
sentence = "What are the symptoms of diabetes?"

# Run model
tokens, attention_weights = compute_self_attention(sentence)

# Convert to DataFrame
df_attention = pd.DataFrame(
    attention_weights,
    index=tokens,
    columns=tokens
)

# Print results
print("Self-Attention Weights:")
print(df_attention)

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df_attention, annot=True, cmap="viridis", fmt=".2f")

plt.title(f"Self-Attention Heatmap: '{sentence}'")
plt.xlabel("Key Tokens")
plt.ylabel("Query Tokens")

plt.tight_layout()
plt.savefig("attention_heatmap.png")
plt.show()

print("Heatmap saved as attention_heatmap.png")
```
